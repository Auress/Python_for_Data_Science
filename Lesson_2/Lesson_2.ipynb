### Задание 3

```python
x = np.linspace(-3,3,51)
x
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x,y1)
ax1.set_title('График y1', fontsize=14)
ax1.set_xlim([-5,5])

ax2.plot(x,y2)
ax2.set_title('График y2', fontsize=14)

ax3.plot(x,y3)
ax3.set_title('График y3', fontsize=14)

ax4.plot(x,y4)
ax4.set_title('График y4', fontsize=14)

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
cr_v1 = creditcard.set_index('Class')['V1']
```

```python
cr_v1.shape
```

```python
c0 = cr_v1[cr_v1.index == 0]
c1 = cr_v1[cr_v1.index == 1]
```

##### Вариант_1

```python
plt.hist(c0, density=True, color = 'grey', label = 'Class 0', 
         bins = 20, alpha = 0.5)
plt.hist(c1, density=True, color = 'red', label = 'Class 1', 
         bins = 20, alpha = 0.5)
plt.xlabel('Class', fontsize=14)
plt.legend()
plt.show()
```

##### Вариант_2 (немножечко измененный график)

```python
colors = ['grey', 'red']
names = ['Class 0', 'Class 1']
plt.hist([c0, c1], density=True, log=True, color=colors, label=names, 
         bins = 20, alpha = 0.7)
plt.xlabel('Class', fontsize=14)
plt.legend()
plt.show()
```
